2. Escreva uma função para verificar se uma palavra é um palíndromo e retorne 1 em caso positivo e 0 (zero) em caso negativo.

````python
def valida_palindromo(palavra):
    '''
    docstring = aparece como comentário de ajuda na função
    '''
    print('palavra',palavra)
    if type(palavra) != str:
        print('Informe uma palavra em string!')
    else:
        num = 0
        palindrome = True
        while num < len(palavra):
            #print(palavra[num])
            #print(palavra[len(palavra)-num-1])
            if palavra[num] != palavra[len(palavra)-num-1]:
                palindrome = False
            num += 1
        if palindrome == False:
            return 0
        else:
            return 1
    
    
result = valida_palindromo(palavra = 'amema')
print(result)
````

6. Monte uma função que vai calcular e comparar os custos de uma viagem para 3 cidades diferentes e que, ao final, irá recomendar qual é o destino mais barato. Essa função deve receber somente uma parâmetro 'noites'. A partir desse argumento a função deve calcular o custo total de viagem para cada uma das cidades. Utilize os dados abaixo.

|   Cidade  | Hotel/noite | Alimentação/noite | Locomoção/noite |
|:---------:|:-----------:|:-----------------:|:---------------:|
| Abu Dhabi |     1300    |        200        |       150       |
|  Genebra  |     1200    |        150        |        80       |
|    Oslo   |     1000    |        250        |       100       |


````python
dict_preços = {}
dict_preços['Abu Dhabi'] = {'hotel':1300, 'alimentação':200, 'locomoção':150}
dict_preços['Genebra'] = {'hotel':1200, 'alimentação':150, 'locomoção':80}
dict_preços['Oslo'] = {'hotel':1000, 'alimentação':250, 'locomoção':100}
dict_total = {}

def calcula_melhor_destino(noites: int):
    for calc in dict_preços:
        #print(calc)
        total = (dict_preços[calc]['hotel'] + dict_preços[calc]['alimentação'] + dict_preços[calc]['locomoção']) * noites
        #print(total)
        dict_total[calc] = total #guarda total das 3 despesas
    
    #loop de forma ordenada no dict, para pegar primeira posição

    
    #transforma o dict em lista, para dar sort e ver qual é o menor preço
    final = list(dict_total.values()) 
    #ordena resultado para pegar menor valor
    final.sort()
    print(final)
    #agora pega a primeira posição (que vai ser o menor valor) 
    menor_valor = final[0]
    #print(menor_valor)
    #Pesquisa qual é a cidade relacionada ao menor valor encontrado
    melhor_destino = [cidade for cidade, valor_total in dict_total.items() if valor_total == menor_valor]
    print('Melhor destino: ',melhor_destino)

#chama função
calcula_melhor_destino(noites=3)
````

11. Construa uma função que receba 3 valores (a,b,c) e retorne uma lista com todos os valores entre 'a' e 'b' que são múltiplos de 'c'. Utilize os valores abaixo (10,500,23).

````python
def acha_mult(lista: tuple):
    lista_mult = []
    print('Parâmetro de Entrada: ',lista)
    a = lista[0]
    b = lista[1]
    c = lista[2]
    for num in range(a,b+1):
        #print(num)
        if num % c == 0:
            #print('multiplo: ',num)
            lista_mult.append(num)
    return lista_mult

lista = acha_mult((10,500,23))
print('Lista de múltiplos: ',lista)
````

17. Considera as duas listas abaixo, a primeira demonstra o valor real de venda de uma casa e a segunda demonstra o valor recomendado por um modelo de regressão. Gere uma terceira lista com o resíduo de cada ponto. O resíduo é o resultado da subtração entre o valor real e o valor predito para determinado ponto.

real = [150358,368525,532152,144789]
predito = [162122,365255,533200,185789]

````python
real = [150358,368525,532152,144789] 
predito = [162122,365255,533200,185789]
#junta as 2 listas
nova_lista = []
for concat in zip(real, predito):
    nova_lista.append(concat)
#calcula resíduo
lista_residuo = []
for soma in nova_lista:
    lista_residuo.append(soma[0]-soma[1])
print(lista_residuo)
````
