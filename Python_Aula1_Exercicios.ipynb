5. Uma escola tem a seguintes regras de avaliação:
* Notas abaixo de 24 -> F
* De 25 a 44 -> E
* De 45 a 49 -> D
* De 50 a 59 -> C
* De 60 a 79 -> B
* Acima de 80 -> B

Crie um algoritmo para avaliar a nota de um aluno atribuida a variável ```nota_aluno```, teste pelo menos 5 notas diferentes.

````python
nota_aluno=float(input('Entre com a nota do aluno: '))
if nota_aluno<=24:
    print('Nota final é F')
elif nota_aluno<=44:
    print('Nota final é E')
elif nota_aluno<=49:
    print('Nota final é D')
elif nota_aluno<=59:
    print('Nota final é C')
elif nota_aluno<=79:
    print('Nota final é B')
else:
    print('Nota final é A')
````

6. A promoção de inverno começou na loja Magazine Luciana, aproveitando essa situação você decidiu comprar um número ```x``` de potes de Toddy, o valor de cada pote é de R\$5.99. A loja te dará um desconto de 10% no valor total se você comprar mais do que 10 potes. Monte um algoritmo que pergunta quantos potes você quer comprar, calcula o valor total (levando em consideração os descontos) e printa o valor a ser pago pelo pedido.

````python
qtde_potes = int(input('Informe quantos potes você deseja comprar: '))
valor_pote = 5.99
valor_total = qtde_potes * valor_pote
if qtde_potes > 10:
    valor_total = valor_total * 0.9
    print(f'Você vai comprar {qtde_potes} potes e irá pagar um total de {valor_total}. Devido à quantidade comprada a loja te deu um desconto de 10%.')
else:
    print(f'Você vai comprar {qtde_potes} potes e irá pagar um total de {valor_total}.')
````

7. Crie um algoritmo que pergunte para o usuário a quantidade de aulas dadas em um semestre, a quantidade de aulas que ele participou. Caso o aluno tenha participado de pelo menos 75% das aulas imprima "Aluno aprovado", em caso negativo, imprima "Aluno reprovado".

````python
qtde_aulas_semestre = int(input('Por favor informe a quantidade de aulas que foram ministradas no semestre: '))
qtde_aulas_assistidas = int(input('E quantas aulas você assistiu? '))
if qtde_aulas_assistidas >= (qtde_aulas_semestre * 0.75):
    print('Aluno aprovado')
else:
    print('Aluno reprovado')
````
