2. Crie uma função que receba duas listas de strings e retorne a concatenação desta lista a nível de elementos.

Exemplo:

lista1 = ['Na','Ma','Jo']

lista2 = ['sser','ria','ão']

saída = ['Nasser','Maria','João']

```python
def concat_lista(lista1: list, lista2: list):
    nova_lista = []
    for id in lista1:
        #print(id)
        #print(lista2[lista1.index(id)])
        nova_lista.append(id+lista2[lista1.index(id)])
    return nova_lista

#chama função
lista = concat_lista( ['Na','Ma','Jo'],['sser','ria','ão'])
print(lista)
```

8. Crie um função que recebe uma lista e retorna o seu desvio padrão.

```python
def calc_desvio(lista):
    ####################################
    #calculando desvio padrão
    ####################################
    n = len(lista) #qtde de elementos na lista
    #------------------------------------
    ma = 0.00 
    #calcula Média Aritmética
    for valor in lista:
        ma = ma + valor
    if n>0:
        ma = ma/n
    else:
        ma = 0
    #print('lista ',lista)
    #print('len ',n)
    #print('Média: ',ma)
    #------------------------------------
    soma = 0 #calcula membro - ma
    for valor in lista:
        soma = soma +  ((valor - ma)**2)
    #print(ma)
    #print(soma)
    if (n-1) > 1:
        desvio = (soma/(n-1))**0.5
    else:
        desvio = 0
    return(round(desvio,2))
    
#chama função
DesvioPadrao = calc_desvio([15,16,17,58,95,68,35,12,12,12,44,87,95,66,78,10,12,73,155,788,157,15,25,77,185,167,142,135,172,111,121,100,101,135,196,117,156,125,189,147,155,178,142,143,156,189,178,112,113,154,144])
print('O desvio padrão calculado é: ',DesvioPadrao)
```

9. SDR (Standard Deviation Reduction) é o cálculo de redução do desvio padrão depois da quebra de um dataset utilizando uma regra. Ele consiste na seguinte fórmula.

sdr.png

Onde :

*   SDR(T,X) é redução do desvio padrão original T dado a regra X
*   S(T) é o desvio padrão original
*   S(T,X) é o desvio padrão ponderado utilizando a regra X.


(use desvio populacional)



exemplo.png

```python
def calc_sdr(aptos,regra):
    #print(aptos)
    aptos_maiorRegra = []
    aptos_menorRegra = []
    for escolha in aptos:
        #print(dict_aptos[escolha])
        if escolha <= regra:
            aptos_menorRegra.append(escolha)
        else:
            aptos_maiorRegra.append(escolha)
    #print(aptos_menor16)
    #print(aptos_maior16)
    #calcula desvio
    if len(aptos_maiorRegra) > 0:
        desvio_maiorRegra = calc_desvio(aptos_maiorRegra)
    else:
        desvio_maiorRegra = 0
    if len(aptos_menorRegra) > 0:
        desvio_menorRegra = calc_desvio(aptos_menorRegra) 
    else:
        desvio_menorRegra = 0
    desvio_total = calc_desvio(aptos)
    print()
    #print('Desvio total: ',desvio_total)
    #print('Desvio aluguéis maiores que 16',desvio_maiorRegra)
    #print('Desvio aluguéis menores que 16',desvio_menorRegra)
    # calcula sdr
    s_MaiorRegra = round(((len(aptos_maiorRegra)/len(aptos))*desvio_maiorRegra) + ((len(aptos_menorRegra)/len(aptos))*desvio_menorRegra),2)
    sdr = round(desvio_total - s_MaiorRegra,2)
    return sdr

#dict_aptos = {'A':15, 'B':20, 'C':35, 'D':45,'E':150, 'F':12, 'G':35,'H':45,'I':10,'J':12,'K':15}
sdr_calc = calc_sdr([15, 20, 35, 45, 150, 12, 35, 45, 10, 12, 15],16)
print(sdr_calc)
```

```python
import matplotlib.pyplot as plt

def plot_sdr(lista_de_sdr):
    f,ax = plt.subplots(1,1,figsize=(16,6))
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)


    ax.set_title('Comportamento de SDR',fontsize=25)
    ax.set_ylabel('SDR',fontsize=15)
    ax.plot(lista_de_sdr)

    plt.show()
```

```python
#chama função
lista_plot = []
for i in range(0,201):
    #print(i)
    sdr_calc = calc_sdr([15,16,17,58,95,68,35,12,12,12,44,87,95,66,78,10,12,73,155,788,157,15,25,77,185,167,142,135,172,111,121,100,101,135,196,117,156,125,189,147,155,178,142,143,156,189,178,112,113,154,144],i)
    #print(sdr_calc)
    lista_plot.append(sdr_calc)
print(lista_plot)

plot_sdr(lista_plot)
```
